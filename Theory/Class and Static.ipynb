class or Static Variabls

```python
#Classic or Static variable--both means the same
class Player:
  total_runs = 0                            #class or static variable
  def __init__(self, name):
    self.name = name
    self.total_runs = 0
  def six(self):
    self.total_runs += 6
    Player.total_runs += 6
  def four(self):
    self.total_runs += 4
    Player.total_runs += 4



print(Player.total_runs)
p1 = Player("Shakib")
p2 = Player("Tamim")
print(p1.total_runs)
p2.six()
p1.four()
p1.six()
print(p1.total_runs)
print(f"{p1.total_runs} + {p2.total_runs} = {Player.total_runs}")

```

Class and Static method

There are 3 types of method:
1. Instance method
2. Class method
3. Static method

```python
class Employee:
  org_name = "Google"
  def __init__(self, name):
    self.name  = name
  @classmethod              #class method declaration  it takes cls reference
  def info(cls):
    return Employee.org_name


Employee.info()
emp1 = Employee("Nahin")
```

```python
class Employee:
  org_name = "Google"
  def __init__(self, name):
    self.name  = name
  @staticmethod              #static method declaration It doesn't take 'cls' unlike class method
  def details():
    print("This is an Employee class")


print(Employee.org_name)
Employee.details()
emp1 = Employee("Nahin")
```

```python
class Student:
  uni_name = "BracU"
  def __init__(self, name, id):
    self.name = name
    self.id = id
  def details(self):
    print("Name:", self.name)
    print("ID:", self.id)
    print("University:", Student.uni_name)

  @classmethod
  def up_uni_name(cls, name):
    cls.uni_name = name

  @staticmethod                    #this is the way to declare a static method
  def check_Department(id):        #this shouldn't have any cls or self reference. It should be  an independent method
    id = str(id)
    if id[0:2]=="23":
      print("CSE")
    elif id[0:2]=="22":
      print("EEE")

s1 = Student("Nahin", 23201416)
s2 = Student("Ayat", 2220141)
Student.check_Department(s1.id)
Student.check_Department(s2.id)
```
