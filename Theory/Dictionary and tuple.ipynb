```python
#assigning a value
my_dict = {
    "aaraf": 7,
    "aayat": 2,
    "nahin": 20
}
print(my_dict)
print(my_dict["aayat"])
print(my_dict.get("aaraf"))
```

```python
#copying the dictionary
my_dict = {
    "aaraf": 7,
    "aayat": 2,
    "nahin": 20
}
another_dict = my_dict.copy()
another_dict["aaraf"] = 8
print(another_dict)
print(my_dict)
```

```python
#some more copying methods
my_dict = {
    "aaraf": 7,
    "aayat": 2,
    "nahin": 20
}
another_dict = dict(my_dict)
print(another_dict)
```

```python
#loop through a dictionary
my_dict = {
    "aaraf": 7,
    "aayat": 2,
    "nahin": 20
}
for i in my_dict:
  print(i, end=" ")
print()
for i in my_dict:
  print(my_dict[i], end=" ")
print()
for i in my_dict:
  print(i, ":", my_dict[i])
```

```python
#more looping through a dictionary
my_dict = {
    "aaraf": 7,
    "aayat": 2,
    "nahin": 20
}
for x, y in my_dict.items():
  print(f"{x}: {y}")
```

```python
#assigning value with for loop
squares = {}
for i in range(6):
  squares[i] = i**2
print(squares)
```

```python
#assigning value with if conditional
ev_squares = {x: x**2 for x in range(11) if x%2==0}
print(ev_squares)
odd_squares = {y: y**2 for y in range(11) if y%2==1}
print(odd_squares)
```

```python
#dictionary membership test
#it checks only key, not value
my_dict = {0: 0, 1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
print(0 in my_dict)
print(6 not in my_dict)
print(25 in my_dict)
```

```python
#methods of dictionary
my_dict = {0: 0, 1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
print(my_dict.items())
print(my_dict.keys())
print()
```

```python
#fromkey method
marks = {}.fromkeys(["math", "science", "english"], 0)
print(marks)
for item in marks.items():
  print(item)
for i, j in marks.items():
  print(f"{i}: {j}")
```

```python
#dictionary built-in function: "sorting" --> it sorts the keys in an ascending order and returns a list
my_dict = {0: 0, 2: 4, 1: 1, 3: 9, 5: 25, 4: 16}
print(sorted(my_dict.keys()))
my_dict = {"math": 20, "science": 40, "english": 90}
print(sorted(my_dict.keys()))
```

```python
#dictionary built-in functions: "all" --> it shows if all the keys are greater than 0 or not
my_dict = {0: 0, 2: 4, 1: 1, 3: 9, 5: 25, 4: 16}
print(all(my_dict))
my_dict = {2: 4, 1: 1, 3: 9, 5: 25, 4: 16}
print(all(my_dict))
```

```python
#dictionary built-in functions: "any" --> it shows wheather any of the keys is greater than 1 or not
my_dict = {0: 0}
print(any(my_dict))
my_dict = {0: 0, 2: 4, 1: 1, 3: 9, 5: 25, 4: 16}
print(any(my_dict))
```

```python
#dictionary built-in functions: "len" --> it shows the length of the dictionary
my_dict = {0: 0, 2: 4, 1: 1, 3: 9, 5: 25, 4: 16}
print(len(my_dict))
```

```python
#dictionary built-in functions: "keys", "values" --> it shows the keys and values of the dictionary respectively
my_dict = {0: 0, 2: 4, 1: 1, 3: 9, 5: 25, 4: 16}
print(my_dict.keys())
print(my_dict.values())
```

```python
#creating a tuple
#empty tuple
my_tuple = ()
print(my_tuple)
#tuple having integers
my_tuple = (1, 2, 3, 4, 5)
print(my_tuple)
#tuple having multiple datatypes
my_tuple = (1, 3, "aayat", 5.5)
print(my_tuple)
```

```python
#nested tuple
my_tuple = (3, 5, [3, 4, 'aaraf'], "nahin", 4.5, (3, 4, 5))
print(my_tuple)
```

```python
#tuple packing and unpacking
#packing
my_tuple = (3, 5.5, "aayat")
print(my_tuple)
#unpacking
a, b, c = my_tuple
print(a)
print(b)
print(c)
```

```python
#To create a tuple with only one item, we have to put a comma after the items otherwise python won't recognise it as a tuple
my_tuple = (3)
print(my_tuple)
my_tuple = (3, )
print(my_tuple)
```

```python
#Accessing elements of tuples
my_tuple = ("a", "a", "y", "a", "t")
print(my_tuple[0])
print(my_tuple[4])
print(my_tuple[-2])
```

```python
#nested tuple accessing
my_tuple = (3, 5, [3, 4, 'aaraf'], "nahin", 4.5, (3, 4, 5))
print(my_tuple[2][2])
print(my_tuple[-1][-3])
```

```python
#mutibility of tuple
#concatenation
my_tuple = (2, 3, 4)
print(my_tuple + (5, 6, 7))
print(my_tuple)
#we can't delete an items from the tuple but we can delete the entire tuple
del my_tuple
```
