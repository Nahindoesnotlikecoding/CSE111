```python
#private instance varible
class Student:
  def __init__(self, name, id):
    self.name = name
    self.__id = id
  def details(self):
    print(f"{self.name}: {self.__id}")
  def set_id(self, id):
    if id>0:
      self.__id = id
    else:
      print("Invalid ID! Enter your ID again.")
  def get_id(self):
    return self.__id
  def set_name(self, name):
    self.name = name
  def get_name(self):
    return self.name


s1 = Student("Nahin", 11)
s2 = Student("Aaraf", 22)
s1.set_id(108271)
s2.set_id(105904)
s1.set_name("carol")
print(s1.get_name())
s2.set_name("bob")
print(s2.get_name())
print(s1.get_id())
s1.details()
s2.details()
```

```python
class ABC:
  def __init__(self, x, y):
    self.x = x
    self.__y = y
  def details(self):
    print("X:", self.x, "&", "Y:", self.__y)
abc1 = ABC(5, 6)
abc2 = ABC(15, 17)
abc1.details()
abc2.details()
```

```python
#private method
class Student:
  def __init__(self, name, id):
    self.name = name
    self.__id = id
  def details(self):
    print(f"{self.name}: {self.__id}")
    self.__method()
  def __method(self):
    print("Private method executed")
s1 = Student("carol", 23)
s1.details()
```
