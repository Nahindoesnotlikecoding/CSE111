```python
class Cars:
  def __init__(self, model, brand, color):
    self.model = model
    self.brand = brand
    self.color = color
  def display(self):
    print(f"This is a {self.color} color {self.brand} car. It's owner's name is {self.model}.")
car1 = Cars("Aaraf", "Porsche", "Blue")
car2 = Cars("Aayat", "Lamborghini", "Red")
car1.display()
car2.display()
```

```python
class familyMembers:
  def __init__(self, name, age):
    self.x = name
    self.y = age
  def display(self):
    print(f"family member's name is {self.x} and his/her age is {self.y}")
p1 = familyMembers("aaraf", 7)
p2 = familyMembers("aayat", 2)
p3 = familyMembers("faria", 28)
p4 = familyMembers("shajib", 37)
p5 = familyMembers("aaraf", 20)
p1.display()
p2.display()
p3.display()
p4.display()
p5.display()
```

```python
class Student:
  def __init__(self, name, id):
    self.x = name
    self.y = id
  def details(self):
    print("Name:", self.x, "ID:", self.y)
s1 = Student("Bob", 11)
s2 = Student("Carol", 22)
s1.details()
s1.y = 24
s1.details()
s2.details()
```

```python
class House:
  def __init__(self):
    self.window = 4
    self.door = 2
  def view(self):
    print(self.window, "Windows", self.door, "doors")
h1 = House()
h2 = House()
h1.view()
h2.view()
h2.door, h2.window = 1, 5
h2.view()
h1.door, h1.window = 3, 6
h1.view()
```

```python
class Home:
  def __init__(self, window, door):
    self.window = window
    self.door = door
  def view(self):
    print(self.window, "Windows", self.door, "Doors")


window = int(input())
door = int(input())
h1 = Home(window, door)
h2 = Home(window+2, door-1)
h1.view()
h2.view()
```

```python
class Dog:
  def __init__(self, name, color):
    self.x = name
    self.y = color
    self.leg = 4
  def display(self):
    print(f"The name of the dog is {self.x}. It is a {self.y} color dog and it has {self.leg} legs.")


d1 = Dog("Tommy", "Redish White")
d2 = Dog("Pinky", "Yellowish Red")
d1.display()
d2.display()
print(d1.__dict__)
print(d2.__dict__)
```

```python
class Dog:
  def __init__(self, name, color):
    self.name = name
    self.color = color
  def update_color(self, color):
    self.color = color
  def poke(self):
    print(self.color, "color dog named", self.name, "is smiling.")

d1 = Dog("Tommy", "Redish White")
d2 = Dog("Pinky", "Yellowish Red")
d1.poke()
d2.poke()
d2.update_color("Brown")
d2.poke()
d1.update_color("White")
d1.poke()
print(d1.__dict__)
print(d2.__dict__)
print(dir(d1))
print(dir(d2))
```

```python
class Book:
  def __init__(self, name, author):
    self.name = name
    self.author = author
    self.price = 0
  def set_price(self, price):
    self.price = price
  def get_price(self):
    return self.price
  def details(self):
    print("Book name:", self.name,
          "\nAuthor:", self.author,
          "\nPrice:", self.price, "Taka")


b1 = Book("Hoimonti", "Humayun Ahmed")
b2 = Book("Opekkha", "Rabindranath Tagore")
b1.details()
b2.details()
print()
b1.set_price(1200)
b2.set_price(1590)
b1.details()
b2.details()
```

```python
#pass-by reference
class Cat:
  def __init__(self, color, action):
    self.color = color
    self.action = action
  def view(self):
    print(self.color, "cat is", self.action)
  def compare(self, cat):
    if self.action == cat.action:
      print("Both cats are", self.action)
    else:
      print("Both cats are different")


c1 = Cat("Blue", "jumping")
c2 = Cat("White", "jumping")
c1.view()
c2.view()
c1.compare(c2)
```

```python
#pass-by value & pass-by reference
class Cat:
  def __init__(self, color, action):
    self.color = color
    self.action = action
  def view(self, num, clr):
    num += 5
    clr1 = clr
    clr1[0] = "Green"
    print("Method inside:", num)
    print("Method inside:", clr)
colors = ["Black", "Red", "Yellow", "Blue"]
c1 = Cat("White", "Jumping")
x = 55
c1.view(x, colors)  #x-->pass-by value    &    colors-->pass-by reference
print("Method outside:", x)
print("Method outside:", colors)
```
