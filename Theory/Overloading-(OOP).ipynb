```python
#One way to achieve method-overloading
class myCalculator:
  def product(self, num1, num2=None, num3=None):
    if num1 != None and num2 != None and num3 != None:
      print(num1 * num2 * num3)
    elif num1 != None and num2 != None:
      print(num1 * num2)
    else:
      print(1*num1)


mul = myCalculator()
mul.product(4)
mul.product(4, 6)
mul.product(4, 6, 5)
```

```python
#Another way to achieve method-overloading
class myCalculator:
  def product(self, *nums):
    total = 1
    for i in nums:
      total *= i
    return total
c1 = myCalculator()
print(c1.product(5, 6, 7, 8))
```

```python
#Another way to achieve method-overloading
from multipledispatch import dispatch
class myCalculator:
  @dispatch(int)
  def product(self, a):
    print(1*a)

  @dispatch(int, int)
  def product(self, a, b):
    print(a*b)

  @dispatch(int, int, int)
  def product(self, a, b, c):
    print(a*b*c)

  @dispatch(int, str, str)
  def product(self, a, b, c):
    print(a*int(b)*int(c))

  @dispatch(int, float, float)
  def product(self, a, b, c):
    print(a*b*c)
c1 = myCalculator()
c1.product(4)
c1.product(4, 5)
c1.product(4, 5, 6)
c1.product(4, "5", "6")
c1.product(4, 5.5, 6.6)
```

```python
#one way of achieving constructor overloading
class Student:
  def __init__(self, *info):
    if len(info) == 3:
      self.name = info[0]
      self.id = info[1]
      self.cgpa = info[2]
    elif len(info)==2:
      self.name = info[0]
      self.id = info[1]
    else:
      self.name = info
    print("Student object created")
s1 = Student("BOB")
s1 = Student("BOB", 23)
s1 = Student("BOB", 23, 3.9)
s1 = Student()
```

```python
#Operator overloading-addition
class data:
  def __init__(self, x):
    self.x = x
  def __add__(self, other):
    return self.x + other.x
num1 = data(10)
num2 = data(20)
str1 = data("CSE")
str2 = data("111")
print(num1+num2)
print(str1+str2)
```

```python
#Operator overloading-greater than
class data:
  def __init__(self, x):
    self.x = x
  def __gt__(self, other):
    if self.x > other.x:
      return True
    else:
      return False
num1 = data(20)
num2 = data(10)
if num1>num2:
  print("num1 is greater")
else:
  print("num2 is greater")
```

```python
#Operator overloading-less than
class data:
  def __init__(self, x):
    self.x = x
  def __lt__(self, other):
    if self.x < other.x:
      return True
    else:
      return False
num1 = data(20)
num2 = data(10)
if num1<num2:
  print("num1 is lesser")
else:
  print("num2 is lesser")
```

```python
#Operator overloading-equal to
class data:
  def __init__(self, x):
    self.x = x
  def __eq__(self, other):
    if self.x == other.x:
      return True
    else:
      return False

num1 = data(20)
num2 = data(21)
if num1 == num2:
  print("Both are equals")
else:
  print("Not equals")
```

```python
class House:
  def __init__(self, w, d):
    self.window = w
    self.door = d
  def view(self):
    print("The house has", self.window, "window(s) and", self.door, "door(s)")
  def __add__(self, other):
    new_window = self.window + other.window
    new_door = self.door + other.door
    return "New house has " + str(new_window) + " window(s) and " + str(new_door) + " door(s)"
h1 = House(6, 2)
h2 = House(4, 1)
print(h1+h2)
```

```python
class House:
  def __init__(self, w, d):
    self.window = w
    self.door = d
  def view(self):
    print("The house has", self.window, "window(s) and", self.door, "door(s)")
  def __add__(self, other):
    new_window = self.window + other.window
    new_door = self.door + other.door
    obj = House(new_window, new_door)
    return obj
h1 = House(6, 2)
h2 = House(4, 1)
h1.view()
h2.view()
h3 = h1 + h2
h3.view()
```
