```python
#1--->there will be a parent/super/base class and some child/sub/derive class
#2--->child class has all the properties of a parent class and also has their own property
#3--->A parent class can't inherit any of the property of a child class
#4--->inheritance comes into picture when a new class possesses the 'IS-A' relationship with an existing class
#5--->let's say B is a child class of A and C is a child class of B. So B has all the properites of A and C has all the properties of B. So, C
#     automatically has all the properties of A
```

```python
#types of inheritance:
#1---> Single inheritance -------> one parent class + one child class
#2---> Multiple inheritance -----------> a child class containg multiple parent class
#3---> Multilevel inheritance --------> a child class which is a child class of an another parent class
#4---> Hybrid inheritance ---------> mixture of everything
#5---> Hierarchical inheritance -----------> one parent class + multiple child class
```

```python
#method resolution order:
#if a method is called it will first check it's own class. if not found then it will go to parent one and other.
```

```python
#single inheritance
class ParentClass:
  def method1(self):
    print("This method is in ParentClass.")
  def method2(self):
    print("This method is in ParentClass.")
class ChildClass(ParentClass):
  def method3(self):
    print("This method is in ChildClass.")


c1 = ChildClass()
c1.method1()
c1.method2()
c1.method3()
```

```python
#single inheritance
#examples
class Animal:
  def __init__(self, name):
    self.name = name
  def eat(self):
    print(self.name, "is eating.")

class Dog(Animal):
  def bark(self):
    print(self.name, "is barking")

a1 = Animal("Jack")
a2 = Dog("Rover")
a2.bark()
a2.eat()
a1.eat()
# a1.bark()
```

```python
#multilevel inheritance
class ParentClass:
  def method1(self):
    print("This method is in ParentClass.")
class ChildClass1(ParentClass):
  def method2(self):
     print("This method is in ChildClass1.")
class ChildClass2(ChildClass1):
  def method3(self):
      print("This method is in ChildClass2.")

c1 = ChildClass2()
c1.method1()
c1.method2()
c1.method3()
```

```python
#multilevel inheritance
#examples
class Student:
  def __init__(self, name, id):
    self.name = name
    self.id = id
  def details(self):
    print("Name:", self.name, "\nID:", self.id)

class CSEStudent(Student):
  def __init__(self, name, id, labs):
    super().__init__(name, id)
    self.labs = labs
  def cry(self):
    print("CSE Students are crying because they have", self.labs, "labs")
class CSEFreshers(CSEStudent):
  def enroll_CSE110(self):
    print(self.name, "has enrolled in CSE-110")


s1 = CSEStudent("Bob", 23201416, 15)
s2 = CSEFreshers("Carol", 23201417, 5)
s2.details()
s1.cry()
s2.enroll_CSE110()
s2.cry()
```

```python
#Hierarchical inheritance
class ParentClass:
  def method1(self):
    print("This method is in parent class")
class ChildClass1(ParentClass):
  def method2(self):
    print("This method is in ChildClass1")
class ChildClass2(ParentClass):
  def method3(self):
    print("This method is in Childclass2")


c1 = ChildClass1()
c2 = ChildClass2()
c1.method1()
c1.method2()
c2.method1()
c2.method3()
```

```python
#Hierarchical inheritance
#examples
class Student:
  def __init__(self, name, id):
    self.name = name
    self.id = id
  def details(self):
    print("Name:", self.name, "\nID:", self.id)

class CSEStudent(Student):
  def __init__(self, name, id, labs):
    super().__init__(name, id)
    self.labs = labs
  def cry(self):
    print("CSE Students are crying because of", self.labs, "labs")
class BBAStudent(Student):
  def party(self):
    print("BBA Students are partying because they have no labs")


s1 = CSEStudent("Bob", 23201416, 15)
s2 = BBAStudent("Carol", 23201417)
s1.details()
s2.details()
s1.cry()
s2.party()
```

```python
#Multiple inheritance
class ParentClass1:
  def method1(self):
    print("This method is in ParentClass1")
class ParentClass2:
  def method2(self):
    print("This method is in ParentClass2")
class ChildClass(ParentClass1, ParentClass2):
  def method3(self):
    print("This method is in ChildClass")

c1 = ChildClass()
c1.method1()
c1.method2()
c1.method3()
```

```python
#Multiple inheritance
#examples
class A:
  def __init__(self):
    print("__init__ of class A")
  def method1(self):
    print("method of class A")
class B:
  def __init__(self):
    print("__init__ of class B")
  def method1(self):
    print("method of class B")
class C(A, B):
  def __init__(self):
    print("__init__ of class C")
  # def __init__(self):
  #   super().__init__()
  #   print("__init__ of class C")
  # def __init__(self):
  #   B.__init__(self)
  def method2(self):
    print("method2 of class C")


c1 = C()
c1.method1()
c1.method2()
B.method1(c1)
```

```python
#Hybrid inheritance
class GrandParentClass:
  def method1(self):
    print("This method is in GrandParentClass")
class ParentClass(GrandParentClass):
  def method2(self):
    print("This method is in ParentClass")
class ChildClass1(ParentClass):
  def method3(self):
    print("This method is in ChildClass1")
class ChildClass2(ParentClass):
  def method4(self):
    print("This method is in ChildClass2")

c1 = ChildClass1()
c2 = ChildClass2()
c1.method1()
c1.method2()
c2.method1()
```

```python
#Super()---> it allows us to call a method of the parent class
class ParentClass:
  def feature1(self):
    print("feature1 is from ParentClass")
  def feature2(self):
    print("feature2 is from ParentClass")
class ChildClass(ParentClass):
  def feature3(self):
    super().feature1()
    print("feature3 is from ChildClass")
    super().feature2()

obj = ChildClass()
obj.feature3()
```

```python
#method overriding
class ParentClass:
  def method1(self):
    print("This method is in ParentClass")
class ChildClass(ParentClass):
  def method1(self):
    print("This method is in ChildClass")

c1 = ChildClass()
c1.method1()
```

```python
#method overriding
class A:
  def __init__(self):
    print("__init__ of class A")
  def method1(self):
    print("olpo study")
  def method2(self):
    print("You will get all my methods")
class B(A):
  def __init__(self):
    # print("__init__ of class B")
    pass
  # def method1(self, num):
  #   print("Always party", num)
  def method1(self):
    super().method1()
    print("Always party")


b = B()
b.method1()
# b.method1(35)
b.method2()
```

```python
#Variable overriding
class Animal:
  def __init__(self, name):
    self.name = name
    self.color = "White"
  def eat(self):
    print(self.color, self.name, "is eating.")

class Dog(Animal):
  def __init__(self, name, color):
    super().__init__(name)
    self.color = color
  def bark(self):
    print(self.color, self.name, "is barking.")


d1 = Dog("Rover", "Brown")
d1.bark()
```

```python
#__str__()
class Student:
  def __init__(self, name, id):
    self.name = name
    self.id = id
    print(self)
  def __str__(self):
    return f"This is {self.name}"

s1 = Student("Bob", 11)
s2 = Student("Carol", 22)
print(s1)
print(s2.__str__())
```
