```python
#keyword-arguments
def childrenName(child1, child2, child3):
  print(f"{child1} is the youngest one.")
  print(f"{child2} is the middle one.")
  print(f"{child3} is the eldest one.")

childrenName("aaraf", "aayat", "Nahin")
```

```python
#keyword-arguments
#This way order of the arguments do not matter
def childrenName(child1, child2, child3):
  print(f"{child1} is the youngest one.")
  print(f"{child2} is the middle one.")
  print(f"{child3} is the eldest one.")

childrenName(child2 = "aayat", child1 = "aaraf", child3 = "Nahin")
```

```python
#multiple arguments
def my_cars(**cars):
  print("The 3rd car's name is", cars["car3"])

my_cars(car1 = "BMW", car2 = "Porsche", car3 = "Lamborghini")
```
