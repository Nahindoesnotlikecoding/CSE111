```python
#Home tasks
#home task-1
def hospital_fee(**fees):
  max = 0
  for i in fees.values():
    if i>max:
      max = i
  max_payer = []
  for i, j in fees.items():
    if j==max:
      max_payer.append(i)
  return max, max_payer

max_amount, max_payer = hospital_fee(Mashrafe = 400, Bumrah = 900, Steyn = 1200, Cummins = 900, Wood = 400, Marsh = 700)
print(f"Highest fee was {max_amount} tk which was paid by", end=" ")
for i in range(len(max_payer)):
  if i==len(max_payer)-1:
    print(max_payer[i], end=".")
  else:
    print(max_payer[i], end=", ")
```

```python
#home task-2
def is_james_bond(lst):
  is_true = False
  for i in range(len(lst)):
    if lst[i]==0:
      for j in range(i+1, len(lst)):
        if lst[j]==0:
          for k in range(j+1, len(lst)):
            if lst[k]==7:
              is_true = True
              break
  return is_true


print(is_james_bond([0, 0, 7, 0, 0, 7, 5]))
```

```python
#home task-3
def assign_students_to_sections(section, *names):
  dct = {}
  sections = []
  for i in section:
    sections.append(i)
  for i in sections:
    dct.update({i: []})
  for i in range(len(names)):
    cur_name = names[i]
    total = 0
    for j in range(len(cur_name)):
      total += ord(cur_name[j])
    x = total % len(sections)
    dct[sections[x]].append(names[i])
  return dct

print(assign_students_to_sections ('ABCDE', 'Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace'))
```

```python
#Assignment-4
def username_generator(first_name, last_name, student_id, middle_name=""):
  username = ""
  if len(first_name)>=3:
    for i in range(3):
      if 65<=ord(first_name[i])<=90:
        username += first_name[i]
      else:
        username += chr(ord(first_name[i])-32)
  else:
    for i in range(len(first_name)):
      if 65<=ord(first_name[i])<=90:
        username += first_name[i]
      else:
         username += chr(ord(first_name[i])-32)
  if middle_name != "":
    username += middle_name
  if len(last_name)>=3:
    for i in range(len(last_name)-3, len(last_name)):
      if 97<=ord(last_name[i])<=122:
        username += last_name[i]
      else:
        username += chr(ord(last_name[i])+32)
  else:
    for i in range(len(last_name)):
      if 97<=ord(last_name[i])<=122:
        username += last_name[i]
      else:
        username += chr(ord(last_name[i])+32)
  username += "_"
  student_id = str(student_id)
  for i in range(len(student_id)-4, len(student_id)):
    username += student_id[i]
  return username

first_name, middle_name, last_name, student_id= input("First Name: "), input("Middle Name: "), input("Last Name: "), int(input("Student ID: "))
print(username_generator(first_name, last_name, student_id, middle_name))
```

```python
#Assignment-5
def key_generator(*names):
  lst = []
  for i in range(len(names)):
    cur_name = names[i]
    key = ""
    if 97<=ord(cur_name[0])<=122:
      key = cur_name[0]
    else:
      key = chr(ord(cur_name[0])+32)
    for j in range(len(cur_name)-2, 0, -1):
      key += str(ord(cur_name[j]))
    if 65<=ord(cur_name[len(cur_name)-1])<=90:
      key += cur_name[len(cur_name)-1]
    else:
      key += chr(ord(cur_name[len(cur_name)-1])-32)
    lst.append(key)
    key = ""
  return f"Encrypted Keys: {lst}"


key_list = key_generator ("Alex", "Bob", "Trudy")
print(key_list)
```

```python
#assignment-6
import random

def playRockPaperScissor(rounds):
  my_score = 0
  com_score = 0
  for i in range(rounds):
      my_inp = input()
      com_inp = random.choice(["rock", "paper", "scissor"])
      print("Computer:", com_inp)
      if my_inp == "rock" and com_inp=="scissor":
          my_score += 1
      elif my_inp == "paper" and com_inp == "scissor":
          com_score += 1
      elif my_inp == "scissor" and com_inp == "rock":
          com_score += 1
      elif my_inp == "rock" and com_inp == "paper":
          com_score += 1
      elif my_inp == "paper" and com_inp == "rock":
          my_score += 1
      elif my_inp == "scissor" and com_inp == "paper":
          my_score += 1
  if com_score > my_score:
      print("Your Score:", my_score)
      print("Computer's Score:", com_score)
      print("Computer has won the game!")
  elif my_score > com_score:
      print("Your Score:", my_score)
      print("Computer's Score:", com_score)
      print("You have won the game!")


rounds = int(input())
playRockPaperScissor(rounds)
```
