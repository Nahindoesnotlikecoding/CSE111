```python
#task 1
class Student:
  def __init__(self, name, cg, credit=9, dept='CSE'):
    self.name = name
    self.cg = cg
    self.credit = credit
    self.dept = dept
    self.scholarship_status = None
  def checkScholarshipEligibility(self):
    if self.cg >= 3.5 and self.credit > 10:
      if self.cg >= 3.7:
        self.scholarship_status = "Merit-based scholarship"
        print(f"{self.name} is eligible for {self.scholarship_status}.")
      elif 3.5 <= self.cg < 3.7:
        self.scholarship_status = "Need-based scholarship"
        print(f"{self.name} is eligible for {self.scholarship_status}.")
    else:
      self.scholarship_status = "No scholarship"
      print(f"{self.name} is not eligible for scholarship.")
  def showDetails(self):
    print("Name:", self.name)
    print("Department:", self.dept)
    print("CGPA:", self.cg)
    print("Number of Credits:", self.credit)
    print("Scholarship Status:", self.scholarship_status)


print('--------------------------')
std1 = Student("Alif", 3.99, 12)
print('--------------------------')
std1.checkScholarshipEligibility()
print('--------------------------')
std1.showDetails()
print('--------------------------')
std2 = Student("Mim", 3.4)
std3 = Student("Henry", 3.5, 15,"BBA")
print('--------------------------')
std2.checkScholarshipEligibility()
print('--------------------------')
std3.checkScholarshipEligibility()
print('--------------------------')
std2.showDetails()
print('--------------------------')
std3.showDetails()
print('--------------------------')
std4 = Student("Bob", 4.0, 6, "CSE")
print('--------------------------')
std4.checkScholarshipEligibility()
print('--------------------------')
std4.showDetails()
```

```python
#task 3
class Department:
    def __init__(self, name="ChE Department", section=5):
        self.dept_name = name
        self.section = section
        self.avg = 0
        print(f"The {self.dept_name} has {self.section} sections.")
    def add_students(self, *students):
        if len(students) == self.section:
            total = 0
            for i in students:
                total += i
            self.avg = total/len(students)
            print(f"The {self.dept_name} has an average of {round(self.avg, 2)} students in each section.")
        else:
            print(f"The {self.dept_name} doesn't have {len(students)} sections.")
    def merge_Department(self, *dept):
        total_student = self.avg * self.section
        for i in dept:
            print(f"{i.dept_name} is merged to {self.dept_name}.")
            total_student += (i.avg * i.section)
        self.avg = total_student / self.section
        return f"Now the {self.dept_name} has an average of {round(self.avg, 2)} students in each section."

d1 = Department()
print('1-----------------------------')
d2 = Department('MME Department')
print('2-----------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------')
d3.add_students(12, 34, 41, 17, 30, 22, 32, 51)
print('6-----------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------')
mega.add_students(21,30,40,36,10,32,27,51,45,15)
print('8-----------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------')
print(mega.merge_Department(d3))
```

```python
#task 4
class Shopidify:
    def __init__(self, name=None):
        self.name = name
        self.item_list = {}
        self.item_history = {}
        self.transaction = 0
        if self.name != None:
            print(f"Welcome {self.name} to Shopidify")
        else:
            print(f"Welcome to Shopidify")
    def add_to_cart(self, item_name, quantity = 1):
        if type(item_name) != list:
            if item_name not in self.item_list:
                self.item_list.update({item_name: quantity})
            else:
                self.item_list[item_name] += quantity
        else:
            for i in range(0, len(item_name), 2):
                x = item_name[i]
                for j in range(i+1, len(item_name)):
                    y = item_name[j]
                    if x not in self.item_list:
                        self.item_list.update({x: y})
                    else:
                        self.item_list[x] += y
                    break
    def display_cart(self):
        if self.name == None:
            print("Items in the cart for Guest:")
        else:
            print(f"Items in the cart for {self.name}:")
        for i, j in self.item_list.items():
            print(f"- {i}: {j}x ")
    def checkout(self):
        self.item_history = self.item_list.copy()
        self.item_list = {}
        self.transaction += 1
        if self.name == None:
            print("Checkout completed for Guest")
        else:
            print(f"Checkout completed for {self.name}")
    def display_history(self):
        if self.name == None:
            print("Purchase history for Guest: ")
        else:
            print(f"Purchase history for {self.name}:")
        print(f"Transaction {self.transaction}: ")
        for i, j in self.item_history.items():
            print(f"- {i}: {j}x ")


guest_account = Shopidify()
print("1xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account = Shopidify("John")
print("2xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan", 2)
guest_account.add_to_cart("Luffy Action Figure")
guest_account.display_cart()
print("3xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.add_to_cart(["Chocolate Chip Cookies", 3,"Goku Action Figure",2,"Dumbbells-5kg",2])
john_account.display_cart()
print("4xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan")
guest_account.display_cart()
print("5xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.checkout()
print("6xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.display_history()
print("7xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.checkout()
print("8xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.display_history()
print("9xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
```

```python
#task 5
class Author:
    def __init__(self, name=None):
        self.name = name
        self.book_list = {}
        self.no_of_books = 0
    def setName(self, name):
        self.name = name
    def addBook(self, book_name, genre):
        if self.name != None:
            self.no_of_books += 1
            if genre not in self.book_list:
                self.book_list.update({genre: [book_name]})
            else:
                self.book_list[genre].append(book_name)
        else:
            print("A book can not be added without author name")
    def printDetail(self):
        print(f"Number of Book(s): {self.no_of_books}")
        print(f"Author Name: {self.name}")
        for i, j in self.book_list.items():
            print(f"{i}: {', '.join(j)}")


a1 = Author()
print("=================================")
a1.addBook("Ice", "Science Fiction")
print("=================================")
a1.setName("Anna Kavan")
a1.addBook("Ice", "Science Fiction")
a1.printDetail()
print("=================================")
a2 = Author("Humayun Ahmed")
a2.addBook("Onnobhubon", "Science Fiction")
a2.addBook("Megher Upor Bari", "Horror")
print("=================================")
a2.printDetail()
a2.addBook("Ireena", "Science Fiction")
print("=================================")
a2.printDetail()
print("=================================")
```
